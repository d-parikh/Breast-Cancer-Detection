#### Dataset can be found at UCI Machine Learning Repository: https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29

<h3>Data Set Information:</h3>

There are 10 predictors, all quantitative, and a binary dependent variable, indicating the presence or absence of breast cancer.
The predictors are anthropometric data and parameters which can be gathered in routine blood analysis.
Prediction models based on these predictors, if accurate, can potentially be used as a biomarker of breast cancer.


<h3>Attribute Information:</h3>
<ul>

## Model Building 

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

### KNearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=5, weights='distance')
KNN.fit(X_train, y_train)
```

```python
y_pred_1 = KNN.predict(X_test)
y_pred_1
```

```python
from scipy.stats import zscore

print('KNeighborsClassifier Algorithm is predicting at : {0:.2g}%'.format(KNN.score(X_test, y_test)*100))
```

### Support Vector Machine

```python
from sklearn.svm import SVC
svc  = SVC(gamma=0.025, C=3)
svc.fit(X_train, y_train)
```

```python
y_pred_2 = svc.predict(X_test)

print('KNeighborsClassifier Algorithm is predicting at : {0:.2g}%'.format(svc.score(X_test, y_test)*100))
```

### Comparing the values

```python
KNNPredictions = pd.DataFrame(y_pred_1)
SVMPredictions = pd.DataFrame(y_pred_2)
predicted_df = pd.concat([KNNPredictions, SVMPredictions], axis=1)
```

```python
predicted_df.columns = [['KNNPredictions', 'SVMPredictions']]
predicted_df
```

### Classification Matrix

```python
from sklearn.metrics import classification_report

print('KNN  Classification Report')
print('..........................')

print(classification_report(y_test, y_pred_1))
```

```python
from sklearn.metrics import classification_report

print('SVC  Classification Report')
print('..........................')

print(classification_report(y_test, y_pred_2))
```

```python
Precision - precision tells how much correctly we were able identify two as two and four as four
Recall - It tells us how true we were able to recall or remember that two is that two and four is actually that four
F1Score - Weighted Average of Precision and Recall
```

### Confusion Matrix

```python
from sklearn import metrics

print('Confusion Matrix for KNN')

cm = metrics.confusion_matrix(y_test, y_pred_1, labels=[2, 4])
df_cm = pd.DataFrame(cm, index = [i for i in[2, 4]], columns=[i for i in ['Predicted B', 'Predicted M']])
sns.heatmap(df_cm, annot=True)
```

From above heatmap, we can see that only 1 were wrongly classiffied as malignant and 5 were wrongly classified as Benign.

While it classified 68 values correctly as Malignant.

```python
print('Confusion Matrix for SVM')

cm = metrics.confusion_matrix(y_test, y_pred_2, labels=[2, 4])
df_cm = pd.DataFrame(cm, index = [i for i in[2, 4]], columns=[i for i in ['Predicted B', 'Predicted M']])
sns.heatmap(df_cm, annot=True)
```

From above heatmap, we can see that only 2 were wrongly classiffied as malignant and 2 were classified as Benign.

While it classified 71 values correctly as Malignant.
